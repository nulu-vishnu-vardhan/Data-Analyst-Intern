```python
import numpy as np
import pandas as pd

dir = "C:\\Users\\N Vishnu\\OneDrive\\Desktop\\KaggleV2-May-2016.csv"

df = pd.read_csv(dir)

print(df.head()) 
```

```python
#dropping null values
df.isnull()
```

```python
#dropping duplicates
df.drop_duplicates()
```

```python
#standardise
col = ['Gender', 'Neighbourhood', 'No-show']
for i in col:
    df[i] = df[i].str.strip().str.lower()
df.head()
```

```python
#date formats
date_col = ['Scheduledday', 'Appointmentday']
for i in date_col:
    df[i] = pd.to_datetime(df[i]).dt.strftime("%d-%m-%y")

df.head()
```

```python
#rename column headers to clean and uniform
df.columns = df.columns.str.strip().str.lower()
df.head()
```

```python
# Check and fixing data types 
df.dtypes
```

```python
df['gender'] = df['gender'].replace('m',1)
df['gender'] = df['gender'].replace('f',0)

df['no-show'] = df['no-show'].replace('no',0)
df['no-show'] = df['no-show'].replace('yes',1)
```

```python
df['patientid'] = df['patientid'].astype(str)
df['appointmentid'] = df['appointmentid'].astype(str)
df['gender'] = df['gender'].astype(int)
df['scheduledday'] = df['scheduledday'].astype(str)
df['appointmentday'] = df['appointmentday'].astype(str)
df['neighbourhood'] = df['neighbourhood'].astype(str)
df['no-show'] = df['no-show'].astype(int)
```

```python
#Cleaned Dataset
df.head(20)
```

```python
#Save
df.to_csv("cleaned_dataset", index = False)
```
